# CONVERTING TO SSS

```python
import math
```

```python
side1 = side2 = side3 = 0 #units
angle1 = angle2 = angle3 = 0 #radians
```

### SAS to SSS
- two sides and angle between the sides given

```python
side1 = int(input("Enter one side: "))
side2 = int(input("Enter other side: "))
angle3 = math.radians(int(input("Enter angle between the sides in degrees: ")))
```

```python
side3 = math.sqrt(side1**2 + side2**2 - 2*side1*side2*math.cos(angle3))
```

```python
# Reordering sides
side1, side2, side3 = sorted([side1, side2, side3])
```

### ASA/AAS to SSS
- Two angle one side given

```python
side = int(input("Enter one side: "))
angle_ = math.radians(int(input("Enter the angle opposite to the side in degrees: ")))
angle__ = math.radians(int(input("Enter another angle in degrees: ")))
```

```python
# Third angle
angle___ = math.pi - angle_ - angle__  # sum of angles = 180°

# Using Law of Sines to find the other sides
# input side is opposite angle_
side_ = side
side__ = side_ * math.sin(angle__) / math.sin(angle_)
side___ = side_ * math.sin(angle___) / math.sin(angle_)
```

```python
side1, side2, side3 = sorted([side_, side__, side___])
```

### Co-ordinates to SSS
- Distance formula
Given:
(x1, y1), (x2, y2), (x3, y3)

```python
print("Enter co-ordinates x,y (comma separated) of 3 points: ")
point1 = input("Enter x1,y1: ").split(",") # list of strings [x1, y1]
point2 = input("Enter x2,y2: ").split(",") 
point3 = input("Enter x3,y3: ").split(",") 

x1,y1 = int(point1[0]), int(point1[1])
x2,y2 = int(point2[0]), int(point2[1])
x3,y3 = int(point3[0]), int(point3[1])
```

```python
print(x1, y1, x2, y2, x3, y3)
```

```python
#Getting the sides
a = math.sqrt((x2 - x3)**2 + (y2 - y3)**2)
b = math.sqrt((x1 - x3)**2 + (y1 - y3)**2)
c = math.sqrt((x1 - x2)**2 + (y1 - y2)**2)
```

```python
# Reordering
side1, side2, side3 = sorted([a,b,c])
```
