# SSS

```python
side1 = int(input("Enter side1: "))
side2 = int(input("Enter side2: "))
side3 = int(input("Enter side3: "))
```

```python
# Sorting in ascending order in order to identify hypotenuse
# using sorted([list])

side1, side2, side3 = sorted([side1, side2, side3])
```

### AREA

```python
s = (side1+side2+side3)/2 #semi perimeter
area = 0
```

```python
# Check for equilateral, Right Angle, if not , use Heron's

if side1 == side2 == side3:
    # equilateral triangle
    area = side1*(math.sqrt(3))/4
    print("Equilateral Triangle")
elif side3*side3 == (side1*side1) + (side2*side2):
    # right angle triangle
    area = 0.5*side1*side2
    print("Right Angled Triangle")
else:
    # heron's formula
    area = (s*(s-side1)*(s-side2)*(s-side3))**0.5
    print("General Triangle")
```

```python
print(area)
```

### TRIANGLE TYPE

```python
triangle_angle_type = ""
triangle_side_type = ""
```

```python
if side1 == side2 == side3:
    triangle_side_type = "equilateral"

elif side1 == side2 or side2 == side3 or side1 == side3:
    triangle_side_type = "isosceles"

else:
    triangle_side_type = "scalene"
```

```python
if side3**2 == side1**2 + side2**2:
    triangle_angle_type = "right"
elif side3**2 <= side1**2 + side2**2:
    triangle_angle_type = "acute"
else:
    triangle_angle_type = "obtuse"
```

```python
print(triangle_angle_type, triangle_side_type)
```
