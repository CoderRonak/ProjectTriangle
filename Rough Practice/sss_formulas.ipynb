### PERIMETERS

```python
def semi_perimeter(a, b, c):
    return (a+b+c)/2 #semi perimeter 's'
```

```python
def perimeter(a, b, c):
    return a+b+c
```

### AREA

```python
def area(a,b,c):
    s = semi_perimeter(a,b,c)
    area = 0

    # Check for equilateral, Right Angle, if not , use Heron's

    if a == b == c:
        # equilateral triangle
        area = a*(math.sqrt(3))/4
    elif c*c == (a*a) + (b*b):
        # right angle triangle
        area = 0.5*a*b
    else:
        # heron's formula
        area = math.sqrt(s*(s-a)*(s-b)*(s-c))
    
    return area
```

### ANGLES

```python
def angles(a,b,c):
    A = B = C = 0
    
    #calculating cos of angles
    cosA = (b**2 + c**2 - a**2)/(2*b*c)
    cosB = (a**2 + c**2 - b**2)/(2*a*c)
    cosC = (a**2 + b**2 - c**2)/(2*a*b)

    #return angles A, B, C in radians
    A = math.acos(cosA)  #cos inverse
    B = math.acos(cosB)
    C = math.acos(cosC)
    
    return A, B, C
    #the input sides will already be sorted hence no need to sort the angles
```

### TRIANGLE TYPE

```python
def triangle_angle_type(a, b, c):
    if c**2 == a**2 + b**2:
        return "right"
    elif c**2 < a**2 + b**2:
        return "acute"
    elif c**2 > a**2 + b**2:
        return "obtuse" # returns string
```

```python
def triangle_side_type(a, b, c):
    if a == b == c:
        return "equilateral"

    elif a == b or b == c or c == a:
        return "isosceles"

    else:
        return "scalene"
```

### INRADIUS, CIRCUMRADIUS

```python
def inradius(a,b,c):
    return area(a,b,c)/semi_perimeter(a,b,c)
```

```python
def circumradius(a,b,c):
    return a*b*c/(4*area(a,b,c))
```

### HEIGHTS (ALTITUDES)

```python
def altitudes(a, b, c):
    area = area(a, b, c)
    
    alt1 = 2*area/a # length of altitude from vertex 1 (angle1) to side 1
    alt2 = 2*area/b
    alt3 = 2*area/c

    return alt1, alt2, alt3 #tuple
```

### MEDIANS (Vertex to Mid of Opposite Side)

```python
def medians(a, b, c):
    m1 = 0.5 * math.sqrt(2*b**2 + 2*c**2 - a**2) #median to mid of side1
    m2 = 0.5 * math.sqrt(2*a**2 + 2*c**2 - b**2)
    m3 = 0.5 * math.sqrt(2*a**2 + 2*b**2 - c**2)

    return m1, m2, m3 #tuple
```
